# error rates

```python
plot_data = data.data['overall_error_rates'].copy()
plot_data[['synthesis', 'GC', 'PCRC', 'Decay', 'Condition', 'Source']] = plot_data.group.str.split('_', expand=True)
plot_data = plot_data.astype({'PCRC': 'int', 'Decay': 'int'})
plot_data = plot_data.pivot(index=['type', 'read', 'synthesis', 'GC', 'PCRC', 'Decay', 'Condition'], columns='Source', values='rate')
plot_data = plot_data.reset_index()
```

```python
fig = px.scatter(
    plot_data,
    x=plot_data["sim"]*1e3,
    y=plot_data["exp"]*1e3,
    color="type",
    hover_data=['synthesis', 'GC', 'PCRC', 'Decay', 'Condition'],
    color_discrete_sequence=["#3182bd", "#756bb1", "#31a354", "#e6550d"]
)


fig.update_layout(
    template='simple_white',
    height=330,
    width=330,
    showlegend=False,
    margin=dict(l=0, r=10, t=10, b=0),
    font_family="Inter",
    legend_font_size=28/3,
)


fig.update_xaxes(
    title='Model error rate / 10<sup>-3</sup> nt<sup>-1</sup>',
    range=[0, 17.5],
    dtick=5,
    minor_ticks="outside", 
    minor_dtick=2.5,
    title_font_family="Inter",
    title_font_size=28/3, 
    tickfont_size=28/3
)
fig.update_yaxes(
    title='Experimental error rate / 10<sup>-3</sup> nt<sup>-1</sup>',
    range=[0, 17.5],
    dtick=5,
    minor_ticks="outside", 
    minor_dtick=2.5,
    title_font_family="Inter",
    title_font_size=28/3, 
    tickfont_size=28/3
)


fig.add_trace(
    pg.Scatter(
        x=np.linspace(0, 100), 
        y=1.2*np.linspace(0, 100),
        line_color="#dddddd",
        line_width=1,
        fill='tonexty', 
        fillcolor="#dddddd",
    )
)
fig.add_trace(
    pg.Scatter(
        x=np.linspace(0, 100), 
        y=0.8*np.linspace(0, 100),
        line_color="#dddddd",
        line_width=1,
    )
)

fig.data = fig.data[::-1]

fig.add_trace(
    pg.Scatter(
        x=np.linspace(0, 100), 
        y=np.linspace(0, 100),
        line_color="#000000",
        line_width=1,
    )
)

fig.add_annotation(
    x=3, y=15,
    text=f"R2: {sklearn.metrics.r2_score(plot_data['exp'], plot_data['sim']):0.2f}<br>MAE: {sklearn.metrics.mean_absolute_error(1e3*plot_data['exp'], 1e3*plot_data['sim']):0.2f}<br>COR: {scipy.stats.pearsonr(plot_data['exp'], plot_data['sim'])[0]:0.2f}",
    showarrow=False,
    align="left",
    font_family="Inter",
    font_size=28/3, 
)


fig.show()
fig.write_image("comparison_errors.svg")
```

# substitution bias

```python
plot_data = data.data['substitutions_by_type'].copy()
plot_data[['synthesis', 'GC', 'PCRC', 'Decay', 'Condition', 'Source']] = plot_data.group.str.split('_', expand=True)
plot_data = plot_data.astype({'PCRC': 'int', 'Decay': 'int'})
plot_data = plot_data.pivot(index=['type', 'read', 'synthesis', 'GC', 'PCRC', 'Decay', 'Condition'], columns='Source', values='rate')
plot_data = plot_data.reset_index()

plot_data[['from_base', 'to_base']] = plot_data["type"].str.split('2', expand=True)

plot_data
```

fill in placeholder data for non-errors (e.g. A2A) so that diagonals are shown in the plot

```python
fillerdf = pd.DataFrame({
    "type": ["A2A", "C2C", "G2G", "T2T"],
    "read": [1, 1, 1, 1],
    "exp": [1.0, 1.0, 1.0, 1.0],
    "sim": [0.0, 0.0, 0.0, 0.0],
    "from_base": ["A", "C", "G", "T"],
    "to_base": ["A", "C", "G", "T"],
})
plot_data = pd.concat([plot_data, fillerdf])
```

```python
fig = px.scatter(
    plot_data,
    x=plot_data["sim"]*1e3,
    y=plot_data["exp"]*1e3,
    color="read",
    facet_col="from_base",
    facet_row="to_base",
    category_orders={'from_base': ["A", "C", "G", "T"], 'to_base': ["A", "C", "G", "T"]},
    hover_data=['synthesis', 'GC', 'PCRC', 'Decay', 'Condition'],
)

fig.update_xaxes(
    title='Model sub. rate / 10<sup>-3</sup> nt<sup>-1</sup>',
    row=1,
)
fig.update_xaxes(
    range=[0, 4],
    dtick=1,
    minor_ticks="outside", 
    minor_dtick=0.5,
    title_font_family="Inter",
    title_font_size=28/3, 
    tickfont_size=28/3
)

fig.update_yaxes(
    title='Exp. sub. rate / 10<sup>-3</sup> nt<sup>-1</sup>',
    col=1
)
fig.update_yaxes(
    range=[0, 4],
    dtick=1,
    minor_ticks="outside", 
    minor_dtick=0.5,
    title_font_family="Inter",
    title_font_size=28/3, 
    tickfont_size=28/3,
)

for i in range(4):
    for j in range(4):
        if i == 3-j: continue
        fig.add_trace(
            pg.Scatter(
                x=np.linspace(0, 10), 
                y=1.2*np.linspace(0, 10),
                line_color="#dddddd",
                line_width=1,
                fill='tonexty', 
                fillcolor="#dddddd",
            ), row=i+1, col=j+1
        )
        fig.add_trace(
            pg.Scatter(
                x=np.linspace(0, 10), 
                y=0.8*np.linspace(0, 10),
                line_color="#dddddd",
                line_width=1,
            ), row=i+1, col=j+1
        )

fig.data = fig.data[::-1]

ix2base = ["A", "C", "G", "T"]
for i in range(4):
    for j in range(4):
        if i == 3-j: continue
        fig.add_trace(
            pg.Scatter(
                x=np.linspace(0, 10), 
                y=np.linspace(0, 10),
                line_color="#000000",
                line_width=1,
            ), row=i+1, col=j+1
        )

        window = plot_data.loc[(plot_data.to_base == ix2base[3-i]) & (plot_data.from_base == ix2base[j])]
        fig.add_annotation(
            x=0.85, y=3.5,
            text=f"R2: {sklearn.metrics.r2_score(window['exp'], window['sim']):0.2f}<br>MAE: {1e3*sklearn.metrics.mean_absolute_error(window['exp'], window['sim']):0.2f}<br>COR: {scipy.stats.pearsonr(window['exp'], window['sim'])[0]:0.2f}",
            showarrow=False,
            align="left",
            font_family="Inter",
            font_size=28/3, 
            row=i+1, col=j+1
        )

fig.update_layout(
    template='simple_white',
    height=660,
    width=660,
    showlegend=False,
    margin=dict(l=0, r=20, t=20, b=0),
    font_family="Inter",
    legend_font_size=28/3,
)

fig.show()
fig.write_image("comparison_substitutions.svg")
```
