```python
fig = make_subplots(rows=1, cols=2, column_widths=[0.5, 0.5], horizontal_spacing=0.05)

for i, idata in enumerate([data1, data2]):
    compare_samples = list(idata.samples.keys())
    compare_samples.remove("0a")

    idata = idata.widedata.copy()

    for ix in compare_samples:
        idata[f"Q_0a-{ix}"] = idata[f"{ix}"]
    idata.replace(np.inf, 0, inplace=True)


    df = idata.copy()
    n_cycles = ['dil. only', 2, 4, 7]

    counts = []
    for cyc in n_cycles:
        factors = np.array(df[f'Q_0a-{cyc}d'] if cyc!= 'dil. only' else df[f'Q_0a-0b'])
        counts.append(factors)

    colors = ["#cccccc", "#bae4b3", "#74c476", "#238b45"]

    for j, (n_cyc, jdata, color) in enumerate(zip(n_cycles, counts, colors)):
        fig.add_trace(
            pg.Violin(
                x=jdata, 
                line_color='#222222',
                line_width=1.5, 
                name=str(n_cyc), 
                fillcolor=color,
                bandwidth=0.075,
            ), row=1, col=i+1,
        )
        std = np.nanstd(np.log(jdata[jdata != 0]))
        fig.add_annotation(x=2.55-0.8*i, y=3-j-0.4,
            text=f"σ = {std:.2f}",
            showarrow=False,
            yshift=0, 
            xanchor="left",
            font=dict(color="#222222", size=28/3),
            row=1, col=i+1,
        )

        
    fig.add_vline(x=1, line_width=1.5, line_dash="dash", line_color="black", row=1, col=i+1)
            
# use negative
fig.update_traces(orientation='h', side='negative', points=False, width=3, opacity=1)
# reverse the (z)-order of the traces
fig.data = fig.data[::-1]
# flip the y axis (negative violin is now positive and traces on the top are now on the bottom)
fig.update_layout(legend_traceorder='reversed')




fig.update_layout(
    template="simple_white", 
    height=330, 
    width=330, 
    showlegend=False, 
    margin=dict(l=0, r=10, t=0, b=0),
    font_family="Inter",
    legend_font_size=28/3,
)
fig.update_xaxes(
    title_text='Normalized coverage',
    title_font_family="Inter", 
    title_font_size=28/3, 
    range=[0, 4],
    dtick=1,
    minor_ticks="outside", 
    minor_dtick=0.5,
    tickfont_size=28/3, 
    row=1, col=1
)
fig.update_xaxes(
    title_text='Normalized coverage',
    title_font_family="Inter", 
    title_font_size=28/3, 
    range=[0, 2.5],
    dtick=0.5,
    minor_ticks="outside", 
    minor_dtick=0.25,
    tickfont_size=28/3, 
    row=1, col=2
)
fig.update_yaxes(
    title_text='Days of storage',
    title_font_family="Inter", 
    title_font_size=28/3, 
    range=[3, -2.0], 
    autorange=False,
    dtick=1,
    tickfont_size=28/3, 
    row=1, col=1
    )

fig.update_yaxes(
    title_text='',
    title_font_family="Inter", 
    title_font_size=28/3, 
    range=[3, -2.0], 
    autorange=False,
    dtick=1,
    tickfont_size=28/3, 
    showticklabels=False,
    row=1, col=2
    )


fig.write_image("bias.svg")
fig.show()
```
